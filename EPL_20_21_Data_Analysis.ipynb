```python
#Import the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sls
%matplotlib inline
```

```python
df_epl = pd.read_csv('EPL_20_21.csv')
```

```python
df_epl.describe()
```

```python
#Getting total number of null values
df_epl.isna().sum()
```

```python
#Create 2 new columns

df_epl['MinsPerMatch'] = (df_epl['Mins'] / df_epl['Matches']).astype(int)
df_epl['GoalsPerMatch'] = (df_epl['Goals'] / df_epl['Matches']).astype(float)
(df_epl.head())
```

```python
#Total Goals

Total_Goals = df_epl['Goals'].sum()
print(Total_Goals)
```

```python
#Penalty Goals
Total_PenaltyGoals = df_epl['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

```python
#Penalty Attempts
Total_PenaltyAttempts = df_epl['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
#Pie Chart for Penalties missed vs scored
plt.figure(figsize=(13, 6))
pl_not_scored = df_epl['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties missed' , 'Penalties Scored']
color = sls.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
#Players from different Nations
np.size((df_epl['Nationality'].unique()))
```

```python
#Most Players from which countries
nationality = df_epl.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(20).plot(kind = 'bar', figsize=(12,6), color = sls.color_palette('magma'))
```

```python
#Clubs with maximum players in their squad

df_epl['Club'].value_counts().nlargest(5).plot(kind = 'bar', color = sls.color_palette('viridis'))
```

```python
#Clubs with Least Players in their squad
df_epl['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color = sls.color_palette('viridis'))
```

```python
#Players based on age group

Under20 = df_epl[df_epl['Age'] <=20]
age20_25 = df_epl[(df_epl['Age'] > 20) & (df_epl['Age'] <=25)]
age25_30 = df_epl[(df_epl['Age'] > 25) & (df_epl['Age'] <=30)]
Above30 = df_epl[df_epl['Age']>30]
```

```python
x = np.array([Under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(),  Above30['Name'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30" ]
plt.title('Total Players With Age Groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```

```python
#Total under 20 Players in each club
players_under_20 = df_epl[df_epl['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind = 'bar' , color = sls.color_palette("cubehelix"))
```
